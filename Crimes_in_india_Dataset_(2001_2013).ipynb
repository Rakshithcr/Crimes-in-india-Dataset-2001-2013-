**Data Visualization**

**STATE WITH MOST CRIMES**

```python
most_crimes_states = data['STATE/UT'].value_counts().head(10)
```

```python
values = most_crimes_states.values
```

```python
index = most_crimes_states.index
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
important_features = [
    "MURDER",
    "ATTEMPT TO MURDER",
    "RAPE",
    "KIDNAPPING & ABDUCTION",
    "ROBBERY",
    "THEFT",
    "HURT/GREVIOUS HURT",
    "DOWRY DEATHS",
    "CRUELTY BY HUSBAND OR HIS RELATIVES",
    "CAUSING DEATH BY NEGLIGENCE",
    "OTHER IPC CRIMES"
]
```

```python
plt.figure(figsize=(18,6))
plt.bar(index,values,label='Crimes')
plt.xlabel('States')
plt.ylabel('Crimes comitted')
plt.title('States with Most Crimes')
plt.legend()
plt.show()
```

**Peaceful States**

```python
peaceful_states = data['STATE/UT'].value_counts().tail(10)
```

```python
peaceful_states_values = peaceful_states.values
peaceful_states_index = peaceful_states.index
```

```python
plt.figure(figsize=(18,6))
plt.bar(peaceful_states_index,peaceful_states_values,label='Crimes')
plt.xlabel('States')
plt.ylabel('Crimes comitted')
plt.title('Peaceful States')
plt.legend()
plt.show()
```

**Numbers of Different Crimes**

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(["MURDER", "RAPE", "THEFT"]):
    plt.subplot(2, 2, i+1)
    data.nlargest(10, feature)[feature].plot(kind='bar', title=feature)
plt.tight_layout()
plt.show()
```

```python
data.nlargest(10, "ATTEMPT TO MURDER")["ATTEMPT TO MURDER"].plot(kind='bar', figsize=(10, 6), title="ATTEMPT TO MURDER")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data.nlargest(10, "HURT/GREVIOUS HURT")["HURT/GREVIOUS HURT"].plot(kind='pie', autopct='%1.1f%%', title="HURT/GREVIOUS HURT")
plt.ylabel('')
plt.show()
```

**Crimes in Different Years**

```python
for feature in important_features[3:]:
    top_10 = data.nlargest(10, feature)
    top_10.plot(x='YEAR', y=feature, kind='bar', figsize=(10, 6), title=f"Top 10 {feature} Cases by Year")
    plt.xlabel('Year')
    plt.ylabel('Number of Cases')
    plt.show()
```

**States with Different Crimes**

```python
for feature in important_features[3:]:
    top_10 = data.nlargest(10, feature)
    top_10.plot(x='STATE/UT', y=feature, kind='bar', figsize=(10, 6), title=f"Top 10 {feature} Cases by States/UT")
    plt.xlabel('States/UT')
    plt.ylabel('Number of Cases')
    plt.show()
```
